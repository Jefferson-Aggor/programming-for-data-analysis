### HANDLING MISSING VALUES


```python
# Count number of missing values
df.isna().sum()
```

```python
NaN_count = sum(df.isna().sum().value_counts().index)

percentage_of_NaN_count = (NaN_count/rows)*100

percentage_of_NaN_count
```

Observations: The percentage of missing values in each column in the dataset is less that 1%

Actions:
Missingness affects only about 0.53% of rows, and comparing means of price/sqft_total/built/grade between rows-with-missing and complete rows shows no major systematic difference.
Listwise deletion is preferred over imputation here, since inventing values for such a small slice of data isn't worth the assumptions it introduces.


```python
# Dropping missing values
df = df.dropna()
print(f"Removed {rows - len(df)} rows with missing values. New shape: {df.shape}")
```

```python
df.isna().sum()
```

### CORRECTING DATA TYPES

1. built(year) was a float because NaNs forced an
   upcast. With no missing values left, it becomes an integer.
2. bedrooms, condition, bathrooms and floors stay float - their .5/.25 values are genuine (half-bathrooms, split levels), not rounding artefacts.
3. All floats are at most 2 decimal places


```python
# Making built an int
df['built'] = df['built'].astype(int)

for col in ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_total', 'floors', 'grade', 'living_area_sqft']:
    df[col] = df[col].round(2)
    

df.head()
```

Resolve the 'renovated' sentinel-value issue
'renovated' mixes a 0 sentinel (never renovated) with real years
(e.g. 1991). Averaging this directly is meaningless.
A binary flag makes it usable in grouping/visuals without distorting statistics.
